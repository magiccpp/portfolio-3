```python
import yfinance as yfin
import pandas as pd
import os
import glob
from sqlalchemy import create_engine, Column, String, Float, DateTime
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from sqlalchemy import PrimaryKeyConstraint  # Import for composite primary key


# --- Database Setup (Same as before, needed to define the table) ---
# Define the connection string
postgres_url = "postgresql+psycopg2://postgres:example@192.168.2.34:5432/app_db"  # replace with your correct URL

# Create the SQLAlchemy engine
engine = create_engine(postgres_url, echo=True)

# Define the declarative base
Base = declarative_base()

# Define the stock_price_training table with date_time and stock_id as primary keys
class StockPriceTraining(Base):
    __tablename__ = 'stock_price_training'

    date_time = Column(DateTime, nullable=False)
    stock_id = Column(String, nullable=False)
    open = Column(Float, nullable=False)
    high = Column(Float, nullable=False)
    low = Column(Float, nullable=False)
    close = Column(Float, nullable=False)
    adj_close = Column(Float, nullable=False)
    volume = Column(Float, nullable=False) # added volume
    currency = Column(String, nullable=False, default="USD")  # Added currency and default value


    __table_args__ = (
        PrimaryKeyConstraint('date_time', 'stock_id'),
    )

    def __repr__(self):
        return (f"<StockPriceTraining(date_time={self.date_time}, stock_id='{self.stock_id}', "
                f"open={self.open}, close={self.close}, high={self.high}, low={self.low}, "
                f"adj_close={self.adj_close}, volume={self.volume}, currency='{self.currency}')>")

# Create the table if it doesn't exist
Base.metadata.create_all(engine)

# --- Data Import Section ---

# Create a session
Session = sessionmaker(bind=engine)
session = Session()

# Define the path to the training data folder
training_data_folder = '../../training_data' # Ensure the path is relative correctly

text_files = glob.glob(os.path.join(training_data_folder, '*.txt'))

for text_file in text_files:
    currency = "USD"  # Default currency
    if 'dax' in text_file:
        currency = "EUR"
    elif 'omx' in text_file:
        currency = "SEK"
    elif 'cac' in text_file:
        currency = "EUR"
    elif 'ftse' in text_file:
        currency = "GBP"

    # read the stock_id from the file, one stock_id per line, and make a list of all stock_ids
    with open(text_file, 'r') as f:
        stock_ids = [line.replace(' ', '-').strip() for line in f.readlines()]
        print(stock_ids)


    # loop through each stock_id and fetch the data then insert the data into the database.
    for stock_id in stock_ids:
        try:
            data = yfin.Ticker(stock_id).history(start='1950-01-01', end=None, auto_adjust=False, timeout=40)
            data.index = pd.to_datetime(data.index)  # Convert index to datetime objects
            # Convert volume to numeric
            data['Volume'] = pd.to_numeric(data['Volume'], errors='coerce')  # Handle potential errors

            if not data.empty:
                for index, row in data.iterrows(): # iterate through dataframe rows
                    # Create a StockPriceTraining object for each row
                    new_entry = StockPriceTraining(
                        date_time=index,
                        stock_id=stock_id,
                        open=row['Open'],
                        close=row['Close'],
                        high=row['High'],
                        low=row['Low'],
                        adj_close=row['Adj Close'],
                        volume=row['Volume'],
                        currency=currency
                    )
                    session.add(new_entry) # add the new entry to the session

                session.commit() # commit changes to the database
                print(f"Successfully imported data for {stock_id}")

        except Exception as e:
            print(f"Error importing data for {stock_id}: {e}")
            session.rollback() # Rollback in case of error


session.close()  # Close the session when done
```
