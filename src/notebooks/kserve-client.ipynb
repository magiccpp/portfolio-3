```python
import requests
import json
import pandas as pd

# Assume df_test_X is your DataFrame containing X_test data
# Convert DataFrame to a list of lists or appropriate format for features
# Example with a single sample (row) from df_test_X

def send_predict_request_http(server_url, model_name, features):
    """
    Sends a prediction request to the KServe HTTP endpoint.

    Args:
        server_url (str): The KServe service endpoint URL (e.g., http://localhost:8080/v1/models/my-kserve-service:predict).
        model_name (str): The name of the model to use, consistent with your data and storage logic
        features (list or list of lists): The input features for prediction.
    """
    payload = {
        "instances": [
            {"model_name": model_name, "features": features}
        ]
    }
    print("Payload:", json.dumps(payload, indent=2))
    headers = {"Content-Type": "application/json"}

    try:
        response = requests.post(server_url, headers=headers, data=json.dumps(payload))

        if response.status_code == 200:
            result = response.json()
            print("Prediction Result:", json.dumps(result, indent=2))
        else:
            print(f"Error: {response.status_code} - {response.text}")

    except requests.exceptions.RequestException as e:
        print(f"Request failed: {e}")

# Example usage (Assuming your service is running on localhost:8080)

# Example 1: single instance
server_url = "http://127.0.0.1:8181/v1/models/dynamic-model-name:predict" # Or your kserve service endpoint

# Replace with the correct column names or order in your DataFrame, or the expected features
#features_row = df_test_X.iloc[0].values.tolist()
#send_predict_request_http(server_url, model_name, features_row)


# Example 2: multiple instances
batch_size = 3 # Number of ROWS to send as batch
features_batch = df_test_X.iloc[0:batch_size].values.tolist() # First batch_size rows
send_predict_request_http(server_url, model_name, features_batch)
```
